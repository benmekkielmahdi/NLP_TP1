## Partie 2 — Nettoyage du texte

```python
import string
text = "Bonjour ! exemple de texte, avec PONCTUATION ?!!!"
text_lower = text.lower()
text_clean = text_lower.translate(str.maketrans('', '', string.punctuation))
text, text_clean
```

```python
print(text_clean)
```

### Réponse
Mise en minuscules = normalisation

## Partie 3 — Tokenisation et stopwords

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

tokens = word_tokenize(text_clean)
stops = stopwords.words('french')
filtered = [w for w in tokens if w not in stops]
tokens, filtered
```

Stopwords = mots sans importance sémantique

## Partie 4 — Lemmatisation

```python
from nltk.stem import WordNetLemmatizer

lemm = WordNetLemmatizer()
lemmatized = [lemm.lemmatize(w) for w in filtered]
filtered, lemmatized
```
